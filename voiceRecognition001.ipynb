```python
import speech_recognition as sr
import pyttsx3
```

```python
# Inicializar SpeechRecognizer
r = sr.Recognizer()

# Función para convertir texto a voz
def SpeakText(command):
    
    #Inicializar el motor
    engine = pyttsx3.init()
    engine.say(command)
    engine.runAndWait()
    
# SpeakText("María Lucía Miguez Buitrago, toca,  el, violín (a veces)")
```

```python
# Convertir texto a voz en el idioma que quiero
from io import BytesIO
import speech_recognition as sr
from gtts import gTTS
from pydub import AudioSegment
from pydub.playback import play

# Usar el micrófono como fuente de entrada
speech = sr.Recognizer()
print('Python ahora está escuchando en Español...')
with sr.Microphone() as source:
    print("Silencio por favor, calibrando ruidos de fondo")
    speech.adjust_for_ambient_noise(source, duration=1)
    while True:
        try:
            print("Calibrado, ahora puede hablar.....")
            # escuchar la entrada del usuario
            audio = speech.listen(source)
            my_input = speech.recognize_google(audio, language="es")
            break
        except sr.UnknownValueError:
            print("No entendí lo que dijo !")
print(f"usted dijo: {my_input}")

    
tts = gTTS(text=my_input,lang="es")
# Crear un archivo temporal
voice = BytesIO()
# Guardar la salida en un archivo de sonido
tts.write_to_fp(voice)
# Reproducir el archivo de voz
voice.seek(0)
play(AudioSegment.from_mp3(voice))
```

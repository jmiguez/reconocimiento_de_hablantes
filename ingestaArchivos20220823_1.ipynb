```python
#-------------------------------------------------------------------------------
#
#
# Ingesta masiva - Parte 2 - Normalización y muestreos
#
#
#-------------------------------------------------------------------------------
import pandas

import os
import subprocess
import librosa
import librosa.display
import sox
from pytube import YouTube
from pathlib import Path
from IPython.display import Audio
from pydub import AudioSegment
from pydub.utils import make_chunks



df = pandas.read_excel("/home/jmmiguez/proyectoAudio/audio/DiputadosDebatenElAcuerdoConElFMI2.xls")
count = 91

for index, row in df.iterrows():
 
    print("segs_ini={} segs_fin={} nombre={}".format(row[0], row[1], row[2]))

    V_SEGS_DESDE=row[0]
    V_SEGS_HASTA=row[1]
    V_NOMBRE_HABLANTE=row[2]

    #-------------------------------------------------------------------------------
    # 7. Normalizar el archivo crudo, usando SOX
    #-------------------------------------------------------------------------------
    V_TFM = sox.Transformer()
    V_TFM.rate(samplerate=16000)
    V_TFM.norm(db_level=-3)
    # Setear recorte del audio entre segundos de inicio y fin
    V_TFM.trim(V_SEGS_DESDE, V_SEGS_HASTA)
    V_TFM.silence(
                location = 0,
                silence_threshold = 0.5,
                min_silence_duration = 0.1,
                buffer_around_silence = False
               )
    # Definir el nombre del archivo de salida normalizado.
    V_PATH_Y_NOMRE_ARCHIVO_NORMALIZADO_WAV = "{}{}_normal.wav".format(C_DATASET_AUDIO_PATH,V_NOMBRE_ARCHIVO_MAESTRO)
    V_TFM.build("{}".format(V_INDEX), "{}".format(V_PATH_Y_NOMRE_ARCHIVO_NORMALIZADO_WAV))
    V_TFM.effects_log

    #-------------------------------------------------------------------------------
    # 8. Librosa: Procesar la señal de audio para ver su amplitud en el tiempo
    #-------------------------------------------------------------------------------
    #plt.figure(figsize=(14, 5))
    #librosa.display.waveshow(V_MUESTRAS, sr=V_SAMPLE_RATE)
    #V_MUESTRAS_NORMAL, V_SAMPLE_RATE_NORMAL = librosa.load(V_PATH_Y_NOMRE_ARCHIVO_NORMALIZADO_WAV)
    #print("cantidad de muestras: {}".format(V_MUESTRAS_NORMAL.shape))
    #print("muestras por segundo: {}".format(V_SAMPLE_RATE_NORMAL))
    #print("duración del audio en segundos: {}".format(V_MUESTRAS_NORMAL.shape[0] / V_SAMPLE_RATE_NORMAL))
    #plt.figure(figsize=(14, 5))
    #librosa.display.waveshow(V_MUESTRAS_NORMAL, sr=V_SAMPLE_RATE_NORMAL)

    #-------------------------------------------------------------------------------
    # 9. Dividir el archivo normalizado, en chunks de 1 segundo.
    #-------------------------------------------------------------------------------
    myaudio = AudioSegment.from_file("{}".format(V_PATH_Y_NOMRE_ARCHIVO_NORMALIZADO_WAV) , "wav") 
    chunk_length_ms = 1000 
    chunks = make_chunks(myaudio, chunk_length_ms) 

    #V_NOMBRE_HABLANTE = input("Indicar el nombre del hablante: ")

    for i, chunk in enumerate(chunks):
        V_PATH_Y_NOMRE_ARCHIVO_WAV = C_DATASET_AUDIO_PATH + V_NOMBRE_HABLANTE + '.wav'
        chunk_name = "{}".format(C_DATASET_AUDIO_PATH) + "{0}.wav".format(i+0)
        #print ("exporting {}".format(chunk_name))
        chunk.export(chunk_name, format="wav")

    #-------------------------------------------------------------------------------
    # 10. Crear un path para el hablante, y mover los archivos allí.
    #-------------------------------------------------------------------------------
    #command = ("rm " + "{}".format(V_PATH_Y_NOMRE_ARCHIVO_MP4))
    #print('command >>',command)
    #subprocess.call(command, shell=True)

    #command = ("rm " + "{}".format(V_PATH_Y_NOMRE_ARCHIVO_WAV))
    #print('command >>',command)
    #subprocess.call(command, shell=True)

    command = ("rm " + "{}".format(V_PATH_Y_NOMRE_ARCHIVO_NORMALIZADO_WAV))
    #print('command >>',command)
    subprocess.call(command, shell=True)

    V_PATH_CHUNK_NORMALIZADO_WAV = C_MAIN_AUDIO_PATH + 'audio/' + V_NOMBRE_HABLANTE
    print('command >> mkdir',V_PATH_CHUNK_NORMALIZADO_WAV)
    os.makedirs(V_PATH_CHUNK_NORMALIZADO_WAV)

    command = ("mv " + "{}".format(C_DATASET_AUDIO_PATH) + "[0-9]*.wav " + V_PATH_CHUNK_NORMALIZADO_WAV)
    print('command >>',command)
    subprocess.call(command, shell=True)
    
    if index == count - 1:
        break
    
print('Fin correcto - Parte 2')
```
